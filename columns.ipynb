```r
df = pd.read_csv("GlobalYouTubeStatistics_Cleaned.csv")
```

```r
get_stats = function(col) {
  list(
    Min = suppressWarnings(min(col, na.rm=True)),
    Max = suppressWarnings(max(col, na.rm=True)),
    Distinct = len(unique(col)),
    Top = names(sort(table(col), decreasing=True))[1]
  )
}
```

```r
get_stats = function(col, name=NULL) {
  col_clean = col[!is.na(col)]
  if (is.numeri[col_clean]) {
    return(list(
      Min = min(col_clean),
      Max = max(col_clean),
      Distinct = len(unique(col_clean))
    ))
  }
  if (!is.null(name) && name == "youtuber") {
    top_vals = df[order(df$rank), "youtuber"][1:3]
    return(list(
      Top = paste(top_vals, collapse=", "),
      Distinct = len(unique(col_clean))
    ))
  }
  freq = sort(table(col_clean), decreasing=True)
  top_vals = names(freq)[1:min(3, len(freq))]
  return(list(
    Top = paste(top_vals, collapse=", "),
    Distinct = len(unique(col_clean))
  ))
}
```

```r
results = list()
for (col_name in names(df)) {
  results[[col_name]] = get_stats(df[[col_name]], col_name)
}
results
```
